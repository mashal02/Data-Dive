```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Load the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()

# Convert to DataFrame
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

```

### First few rows of the dataset

```python
print(iris_df.head())
print(iris_df.shape)
```

# Exercise 1: Data Cleaning




## Handling missing values


```python
print("Number of missing values in each column:")
print(iris_df.isnull().sum())
```

## Removing duplicates


```python
print("\nNumber of duplicate rows:")
print(iris_df.duplicated().sum())
print(iris_df.shape)
iris_df=  iris_df.drop_duplicates()
print(iris_df.shape)
```


# Exercise 2: Data Visualization
## Histogram


```python
print("Histogram of Sepal Length (cm):")
plt.figure(figsize=(8, 6))
sns.histplot(data=iris_df, x='sepal length (cm)')
plt.title('Histogram of Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```


## Boxplot


```python
print("\nBoxplot of Petal Width (cm) by Species:")
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris_df, x='target', y='petal width (cm)')
plt.title('Boxplot of Petal Width (cm) by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.xticks(ticks=[0, 1, 2], labels=iris.target_names)
plt.show()
```


## Scatter plot



```python
print("\nScatter plot of Sepal Length (cm) vs. Sepal Width (cm):")
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='sepal length (cm)', y='sepal width (cm)', hue='target', palette='Set2', s=100)
plt.title('Scatter plot of Sepal Length (cm) vs. Sepal Width (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```
